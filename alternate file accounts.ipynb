```python
a=input('choose create account or login: ')
if(a=='create account'):
    check=open("a&p.txt","r")
    user=input("enter username: ")
    d=0
    for eline in check:
        line=eline.split(",")
        z=line[0].rstrip()
        if(user!=z):
            d=0
        else:
            d=1
            break
    if(d!=0):
        print("account already exists")
    else:
        file=open("a&p.txt","a")
        g=input("enter password: ")
        h=input("confirm password: ")
        while(g!=h):
            h=input("confirm password: ")
        else:
            file.write("\n")
            file.write(user+","+g)
            print("account created")
        file.close()
    check.close()
elif(a=='login'):
    b=open('a&p.txt','r')
    c=input('enter username: ')
    d=0
    g=0
    h=0
    for zline in b:
        g+=1
        z=zline.split(",")
        e=z[0].rstrip()
        if(c==e):
            h=g
            d=1
            break
        else:
            d=0
    if(d):
        f=open("a&p.txt",'r')
        g=input("enter password: ")
        i=0
        for yline in f:
            y=yline.split(",")
            x=y[1].rstrip()
            k=str(x)
            i+=1
            if(i==h):
                if(g==k):
                        print("login accepted")
                elif(g!=k):
                    j=2
                    while(j):
                        print(j,"attempts remaining")
                        g=input("enter password: ")
                        j-=1
                        if(g==k):
                            print("login accepted")
                            break
                    else:
                        print("try later")
                
        f.close()
    else:
        print("account not found")
    b.close()
```
